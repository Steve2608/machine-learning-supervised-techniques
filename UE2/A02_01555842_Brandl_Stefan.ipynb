## Task 1
Consider a one-dimensional classification problem with $X = \mathbb{R}$ and $Y = \{−1, +1\}$. The marginal distribution of labels is given as follows:

\begin{align*}
p(y=-1)&=\frac{3}{4} & p(y=+1)&=\frac{1}{4}
\end{align*}

The conditional distribution $p(x \mid y)$ is given in the following way:

\begin{align*}
p(x \mid y=-1) &= \frac{1}{2\sqrt{\pi}}\exp\left(-\frac{(x+1)^2}{4}\right) &&\text{that is }\mathcal{N}\left(\mu_-=-1,\sigma_-^2=2\right) \\
p(x \mid y=+1) &= \frac{\sqrt{2}}{\sqrt{\pi}}\exp\left(-2(x-3)^2\right) &&\text{that is }\mathcal{N}\left(\mu_+=3,\sigma_+^2=\frac{1}{4}\right)
\end{align*},

where subscripts $-/+$ indicate class memberships.
Visualize the marginal distribution $p(x)$ in one plot and the two conditional distributions $p(y = −1 \mid x)$ and $p(y = +1 \mid x)$ together in another plot. Guess from the visualization what the Bayes-optimal classifier is like.

## Task1 &mdash; Solution
Your analytic results go here.

```python
from matplotlib import pyplot as plt
from numpy import pi as PI, sqrt, exp, array, arange

def negative(x):
    return 1/(2*sqrt(PI)) * exp(((x+1)**2) / -4)

def positive(x):
    return sqrt(2)/sqrt(PI) * exp(-2*(x-3)**2)


x = arange(-6, 6, 0.01)
plt.title("Conditional distributions")
plt.xlabel("x")
plt.ylabel("p(x|y)")
line_neg, = plt.plot(x, negative(x) * 3/4)
line_pos, = plt.plot(x, positive(x) * 1/4)
plt.legend((line_neg, line_pos), ("negative", "positive"))
plt.show()

plt.title("Marginal distribution")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.plot(x, negative(x) * 3/4 + positive(x) * 1/4)
plt.show()
```

From the visualization I guess that the Bayes-optimal classifier is slightly below $2$.

## Task 4
Perform polynomial regression to illustrate the bias-variance decomposition of Task 3. To this end, perform the following steps for each degree $n = 1, \dots, 12$:
 1. Create 200 training sets with $l=20$ samples each.
 2. For each of the training sets, train a polynomial model with degree $n$ and compute the predicted value for $x_0 = 1.8$.
 3. Estimate the squared bias and the variance from the 200 predicted values and compute an overall estimate for the expected prediction error for $x_0 = 1.8$.
 
After having followed these steps, visualize your results appropriately. Discuss how the results illustrate the bias-variance decomposition.

## Task 4 &mdash; Solution

```python
import matplotlib.pyplot as plt

import numpy as np

from numpy import sqrt, zeros, mean, var, linspace
from numpy.random import normal as qnorm, sample, seed

def f(x):
    return 0.6*x**4 + 2*x**3 - 8*x**2

def f_noise(x):
    return f(x) + qnorm(0, sqrt(0.09), len(x))

# for reproducibility
seed(17)

lower, upper = -1, 3
n_samples = 200
l = 20
n = list(range(1, 13))
x0 = 1.8

predictions = zeros((len(n), n_samples))
x_vals = linspace(lower - 1, upper + 1, 50)

for k in n:
    for i in range(n_samples):
        x = sample(l) * (upper - lower) + lower
        y = f_noise(x)
        
        poly = np.polyfit(x, y, k)
        regress_function = np.poly1d(poly)
        predictions[k - 1, i] = regress_function(x0)
        
        x_vals = linspace(lower - 1, upper + 1, 50)
        plt.plot(x_vals, regress_function(x_vals), color="k", alpha=0.05)
        
    plt.title(f"{k}-grade regression")
    plt.ylim((-20, 30))
    plt.plot(x_vals, f(x_vals), color="r", linewidth=2)
    plt.axvline(x=x0, color="b")
    plt.show()

    
for i, line in zip(range(len(n)), predictions):
    print("[n={:2}] MSE:  {:8.4f}  MEAN: {:8.4f}  VAR: {:8.4f}".format(i + 1, ((line - f(x0))**2).mean(), line.mean(), line.var()))
  
plt.title("Bias squared")
plt.plot(n, [((predictions[i - 1, :] - f(x0))**2).mean() for i in n], color="b")
plt.xlabel("n")
plt.show()
```

### Discussion
We can see that very low $n$s leads to underfitting - the original polynomial function is of grade 4 (but looks like grade 3), so we need $n \ge 3$ in order to approximate the function well. 

Very large $n$s lead to overfitting - the samples of the original polynomial function are overfitted grossly, as we can see in "12-grade regression". One outlier alone skews the results immensely.
